```python
def greet(text): 

    print(f"Shout has been called with input {text}")
    return text.upper()

greet("example text")

```

```python
import gradio as gr 
```

```python
# Adding inbrowser=True opens up a new browser window automatically

simple_chat = gr.Interface(fn=greet, inputs="textbox", outputs="textbox", flagging_mode="never")
```

```python
simple_chat.launch(share=True, auth = ("Dzi", "dzi123"), inbrowser=True)
```
