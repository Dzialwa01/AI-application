```python
import os
from dotenv import load_dotenv
from openai import OpenAI
```

```python
load_dotenv(override=True)
openai_api_key = os.getenv('OPENAI_API_KEY')

client = OpenAI (api_key = openai_api_key)
```

```python
def get_response(prompt):

    response = client.chat.completions.create(

        model = "gpt-4.1-mini",
        messages = [{"role": "user", "content": prompt}]
    )

    return response.choices[0].message.content
```

```python
# Exercise 1 -  Use Case: Market Research Report Summarization

# A market research firm receives very long reports about: Market trends, customer behaviour and new technologies
# They are specifically interested in Artificial Intelligence, Data privacy and how these two are affecting customers
# Instead of reading the full report everytime, they want to use AI summarization

Report = """The market has seen rapid growth in the adoption of artificial intelligence across industries such as finance, healthcare, and retail. Businesses are increasingly using AI to personalize customer experiences, automate customer support, and analyze large datasets for better decision-making. However, this rapid adoption has raised concerns among customers regarding data privacy and security.

Many consumers are becoming more cautious about how their personal data is collected, stored, and usedby AI-driven systems. Regulatory frameworks and data protection laws are influencing how companies design their AI solutions. As a result, companies that prioritize transparency and ethical data practices are gaining higher customer trust and loyalty."""

prompt = f"""
Summarize the following market research report in a maximum of five sentences points.
Focus specifically on how artificial intelligence and data privacy are shaping the market and how they are affecting customer behaviour

Here is the report: {Report}
"""

print("Summarized report: \n", get_response(prompt))
```

```python
# Exercise 2 - Use Case: Product Features Summarization (Smartphone)

# An electronics review website wants to: - Show short and clear product feature summaries, help users compare smartphones quickly and avoid long paragraphs.
# So, they decide to use AI summarization to convert long product descriptions into bullet points.

product_description = """The smartphone features a 6.7-inch AMOLED display with a 120Hz refresh rate, offering smooth visuals and vibrant colors. It is powered by a high-performance processor that ensures fast multitasking and gaming. The device includes a 5000mAh battery that supports fast charging, allowing extended usage throughout the day.

The smartphone comes with a triple-camera system, including a 64MP main camera, an ultra-wide lens,and a macro sensor, delivering high-quality photo sand videos. It also offers 5G connectivity, enhanced security features such as an in-display fingerprint sensor, and runs on the latest version of the operating system."""

prompt2 = f"""
Summarize the following smartphone product descrription in no more than five bullet points
Highlight the key features that would help users to quickly compare and evaluate the product

product description: {product_description}"""

print("Original description: \n", product_description)
print("Summarized description: \n", get_response(prompt2))
```

```python
# Exercise 3: Product description expansion

# As you continue your work on the electronics review website, you've come across some products that are already summarized but lack a comprehensive description. Your task now is to expand these concise product descriptions into detailed narratives, ensuring that each product has both a full description and a bulleted summary for easy comparison. The complete description should effectively capture the product's unique features, benefits, and potential applications. You will apply your first prompt on a smart home security camera's product description.
# Craft a prompt that expands the 'product_description` pre-loaded string, and writes a one paragraph comprehensive overview capturing the key information of the product: unique features, benefits, and potential applications.

prompt3 = f"""
Expand The product description and write a one paragraph comprehensive overview of the product.
In the paragrapg include the unique features, benefits, and potential applications

Product description: {product_description}"""

print("Original description: \n", product_description)
print("Expanded description: \n", get_response(prompt3))
```
