```python
# Exercise 1: Extracting text with PyPDF

# - Import the required class from `pypdf` and use it to load the PDF file.
# - Access each page and extract its content using the correct method.
```

```python
from pypdf import PdfReader
```

```python
# Extract text from the PDF
# reader =  PdfReader(r"C:\Users\Dzialwa\OneDrive - Velisa Africa\Github\AI application\AI-application\HR_Employee_Handbook_Sample.pdf")
reader = PdfReader("HR_Employee_Handbook_Sample.pdf")
# Extract text from all pages
document_text = ""
for page in reader.pages: 
    document_text += page.extract_text()

print(document_text)
```

```python
# Exercise 2: Building a Q&A pipeline

# - Initialize a question-answering pipeline with the correct task and model.
# - Pass the question and `document_text` to the pipeline.
# - Print the result to view the answer
```

```python
from transformers import ( 
    AutoTokenizer,
    AutoModelForQuestionAnswering
    )
import torch
```

```python
model_name = "distilbert-base-uncased-distilled-squad"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForQuestionAnswering.from_pretrained(model_name)

# document used is the sample document from above exercise

question = "How many volunteer days are allowed per year?"

inputs = tokenizer(
    question,
    document_text,
    return_tensors = "pt"
)

with torch.no_grad():
    outputs = model(**inputs)

answer_start = torch.argmax(outputs.start_logits)
answer_end = torch.argmax(outputs.end_logits) + 1

answer = tokenizer.convert_tokens_to_string(
    tokenizer.convert_ids_to_tokens(
        inputs["input_ids"][0][answer_start:answer_end]
    )
)   

print("Answer:", answer)
```
